### Example 5.0 -- Left Join
We want to combine the SAP df with the new_df we just imported. We want to keep all of the records that we pulled from SAP and just add the columns from new_df to the records where it matches up. 

```python
# SANITY CHECK
## Since we want to keep all of the records in the SAP dataset, we will be looking for the final df to have 25695 rows. The only thing that should change after the merge is the number of columns. 
sap_df.shape
```

```python
new_df.shape
```

```python
## NOTE: if the primary key column has the same name in both data sets, you don't need to do left_on or right_on. you can just do on = "insert_column_name_here".
left_join_df = pd.merge(sap_df,new_df,how='left',left_on='SSO', right_on='SSO number')
```

### Example 6.0 -- Right Join

We want to combine the SAP df with the new_df we just imported. We want to keep all of the records from new_df and just add the columns from sap_df to the records where the primary key matches up. 

same prompt as before. we want the number of rows to equal 25695. Left join and Right joins are effectively the same thing

In the merge we did above, I put sap_df first in the pd.merge statement and new_df second. This means that I assigned sap_df as the 'left' dataset and new_df as the right. 
We can flip this order and to a right merge and achieve the same results.

```python
right_join_df = pd.merge(sap_df,new_df,how='right', left_on='SSO',right_on='SSO number')
```

```python
right_join_df.shape
```

### Example 7.1 -- Inner Join

#### Data Exploration

Question: How many of these employees are still in DTC?

```python
# checking out the data after the merge
inner_join_example_2.head()
```

Since BUSINESS_SEGMENT_TRENDING is a column that exists in both sap_df and dtc_df, python assigns one as BUSINESS_SEGMENT_TRENDING_x and the other as BUSINESS_SEGMENT_TRENDING_y. The x and y are decided based on the order that you put the dataframes in the merge. So, in this case, BUSINESS_SEGMENT_TRENDING_x is the columns from sap_df and BUSINESS_SEGMENT_TRENDING_y is coming from dtc_df. 

To see how many employees are still in DTC, we want to look at BUSINESS_SEGMENT_TRENDING_x because that comes from the dataset with the more up to date information (sap_df) 

```python
# to figure this out, we want to filter for people who are in direct to consumer in the more current dataset (AKA sap_df)
## then we want to count the unique number of SSOs that match that criteria
inner_join_example_2['SSO'][inner_join_example_2['BUSINESS_SEGMENT_TRENDING_x'] == 'DIRECT-TO-CONSUMER'].nunique()
```

```python
## just curious... where are the people now that used to be in DTC and are not anymore? 
inner_join_example_2[inner_join_example_2['BUSINESS_SEGMENT_TRENDING_x'] != 'DIRECT-TO-CONSUMER']
```
